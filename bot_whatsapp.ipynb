```python
#Abre o navegador Chrome e acessa o WhatsApp Web.
#Espera 10 segundos para o usuário escanear o QR Code.

driver = webdriver.Chrome()
driver.get('https://web.whatsapp.com/')
time.sleep(10)
```

```python
#Usa ActionChains para simular um clique mais preciso na conversa.
funcao_mouse = webdriver.common.action_chains.ActionChains(driver)
funcao_mouse.move_to_element_with_offset(mouse,0,-20)
funcao_mouse.click().perform()
funcao_mouse.click().perform()
```

```python
#Lê o número de telefone do contato da conversa.
ler_cel_franquia = driver.find_element_by_class_xpath('//*@id="main"]/header/div[2]/div[1]/div/span')
celular_franquia = ler_cel_franquia.text
print(celular_franquia)

#Lê as mensagens da conversa.
#Pega a última mensagem recebida.
ler_mensagens = driver.find_elements_by_class_name('_1Gy50')
lista_mensagens =[e.text for e in ler_mensagens]
ultima_mensagem = lista_mensagens[-1]
print(ultima_mensagem)
```

```python

#Encontra o campo de digitação da mensagem.
#Envia a resposta recebida do servidor.

resposta_bot = driver.find_elements_by_xpath('//*[@id="main"]/footer/div[1]/div[2]/div/div[2]')
resposta_bot.click()

banco_resposta = requests.get("http://localhost/bot/index.php", params={'mensagem': ultima_mensagem, 'celular': celular_franquia})
bot_resposta = banco_resposta.text

#Envia a última mensagem + número para um servidor local PHP.
#Recebe uma resposta automatizada com base nesses dados.

time.sleep(3)
resposta_bot.send_keys(bot_resposta, Keys.ENTER)
```

```python
import requests
import pandas as pd


def fetch_all_data(cod_cli_sponte, data_inicial, data_final, api_key='fad1f084-3ce9-4041-9b2c-7fe446447b21'):
    url = 'https://sponte-bi.sponteweb.com.br/api/v1/extracoes/fatoalunosativos'

    headers = {
        'x-api-key': api_key
    }

    page_number = 1
    page_size = 10
    all_items = []

    while True:
        params = {
            'CodCliSponte': cod_cli_sponte,
            'DataInicial': data_inicial,  # Supondo que a API aceite DataInicial e DataFinal
            'DataFinal': data_final,
            'PageNumber': page_number,
            'PageSize': page_size
        }

        try:
            response = requests.get(url, headers=headers, params=params)
            response.raise_for_status()
            data = response.json()

            if not data['items']:
                break

            all_items.extend(data['items'])
            page_number += 1

            if not data['hasNext']:
                break

        except requests.exceptions.HTTPError as err:
            print(f'HTTP error occurred: {err}')
            break
        except Exception as e:
            print(f'An error occurred: {e}')
            break

    return all_items

def export_to_excel(data, filename='dados.xlsx'):
    # Cria um DataFrame a partir dos dados
    df = pd.DataFrame(data)

    # Exporta para Excel
    df.to_excel(filename, index=False)
    print(f'Dados exportados para {filename}')

# Exemplo de uso
if __name__ == '__main__':
    cod_cli_sponte = 422101  # Exemplo de valor
    data_inicial = '2023-01-01T00:00:00Z'  # Data inicial
    api_key = 'fad1f084-3ce9-4041-9b2c-7fe446447b21'  # Defina a chave da API aqui, se necessário

    dados = fetch_all_data(cod_cli_sponte, data_inicial, data_final)

    if dados:  # Verifica se há dados para exportar
        export_to_excel(dados)
    else:
        print('Nenhum dado encontrado para exportar.')
```
